#### **3.4.1 Desviación Estándar**

La **desviación estándar** es una medida que nos indica qué tan dispersos o alejados están los valores de un conjunto de datos respecto a su media o promedio. En otras palabras, nos ayuda a entender cuánto varían los valores individuales de un conjunto de datos con respecto a su valor promedio.

<br>

Imagina que tienes un conjunto de datos que representa las alturas de estudiantes en una clase. Si la desviación estándar es baja, significa que la mayoría de los estudiantes tienen alturas muy similares entre sí y cercanas a la altura promedio de la clase. Por otro lado, si la desviación estándar es alta, significa que las alturas de los estudiantes están muy dispersas, algunos son muy altos y otros muy bajos en comparación con la altura promedio.

<br>

Entonces, la desviación estándar nos da una idea de la dispersión o variabilidad de los datos alrededor de su promedio. Una desviación estándar baja indica que los datos están más agrupados alrededor del promedio, mientras que una desviación estándar alta indica que los datos están más dispersos o separados del promedio.

<br>

En este caso, todas las columnas numéricas tienen desviaciones estándar diferentes de cero, lo que indica que no tienen un único valor.

#### **4.2.1 Renombrar Datos**

**Renombrar datos usando `loc[]`:**

```python
# Identificar las columnas a remover (6)
antes = data['Trabajo'].value_counts().to_frame()

# Renombrar los datos
data.loc[data['Trabajo'] == 'admin.', 'Trabajo'] = 'admin'
data.loc[data['Trabajo'] == 'administrative', 'Trabajo'] = 'admin'
data.loc[data['Trabajo'] == 'Management', 'Trabajo'] = 'management'
data.loc[data['Trabajo'] == 'MANAGEMENT', 'Trabajo'] = 'management'
data.loc[data['Trabajo'] == 'Self-employed', 'Trabajo'] = 'self-employed'
data.loc[data['Trabajo'] == 'Retired', 'Trabajo'] = 'retired'
data.loc[data['Trabajo'] == 'Services', 'Trabajo'] = 'services'

# Dataframe modificado
despues = data['Trabajo'].value_counts().to_frame()

# Visualizar la comparación
display(HTML(html.format('Antes', antes.to_html(), 'Después', despues.to_html())))
```

```python
# Identificar las columnas a remover (3)
antes = data['Estado Civil'].value_counts().to_frame()

# Renombrar los datos
data.loc[data['Estado Civil'] == 'div.', 'Estado Civil'] = 'divorced'
data.loc[data['Estado Civil'] == 'DIVORCED', 'Estado Civil'] = 'divorced'
data.loc[data['Estado Civil'] == 'Single', 'Estado Civil'] = 'single'

# Dataframe modificado
despues = data['Estado Civil'].value_counts().to_frame()

# Visualizar la comparación
display(HTML(html.format('Antes', antes.to_html(), 'Después', despues.to_html())))
```

**Renombrar datos usando `np.select`:**

```python
# Identificar las columnas a remover (6)
antes = data['Educacion'].value_counts().to_frame()

# Renombrar los datos
condiciones = [
    (data['Educacion'] == 'UNK'),
    (data['Educacion'].isin(['SECONDARY', 'Secondary', 'sec.'])),
    (data['Educacion'] == 'Primary'),
    (data['Educacion'] == 'Tertiary')
]
valores = ['unknown', 'secondary', 'primary', 'tertiary']
data.loc[:, 'Educacion'] = np.select(condiciones, valores, default=data['Educacion'])

# Dataframe modificado
despues = data['Educacion'].value_counts().to_frame()

# Visualizar la comparación
display(HTML(html.format('Antes', antes.to_html(), 'Después', despues.to_html())))
```

```python
# Identificar las columnas a remover (4)
antes = data['Prestamo'].value_counts().to_frame()

# Renombrar los datos
condiciones = [
    (data['Prestamo'].isin(['YES', 'Yes'])),
    (data['Prestamo'].isin(['NO', 'No']))
]
valores = ['yes', 'no']
data.loc[:, 'Prestamo'] = np.select(condiciones, valores, default=data['Prestamo'])

# Dataframe modificado
despues = data['Prestamo'].value_counts().to_frame()

# Visualizar la comparación
display(HTML(html.format('Antes', antes.to_html(), 'Después', despues.to_html())))
```

**Renombrar datos usando `np.where()`:**

```python
# Identificar las columnas a remover (2)
antes = data['Contacto'].value_counts().to_frame()

# Renombrar los datos
data.loc[:, 'Contacto'] = np.where(data['Contacto'] == 'phone', 'telephone', data['Contacto'])
data.loc[:, 'Contacto'] = np.where(data['Contacto'] == 'mobile', 'cellular', data['Contacto'])

# Dataframe modificado
despues = data['Contacto'].value_counts().to_frame()

# Visualizar la comparación
display(HTML(html.format('Antes', antes.to_html(), 'Después', despues.to_html())))
```

```python
# Identificar las columnas a remover (2)
antes = data['Resultado Previo'].value_counts().to_frame()

# Renombrar los datos
data.loc[:, 'Resultado Previo'] = np.where(data['Resultado Previo'] == 'UNK', 'unknown',
                   np.where(data['Resultado Previo'] == 'Success', 'success', data['Resultado Previo']))

# Dataframe modificado
despues = data['Resultado Previo'].value_counts().to_frame()

# Visualizar la comparación
display(HTML(html.format('Antes', antes.to_html(), 'Después', despues.to_html())))
```

## **7. APRENDIZAJE AUTOMÁTICO**

### **7.1 Importar Librerías**

```python
# Importar librería scikit-learn, ya que es una de las librerías más populares para Machine Learning

from sklearn.model_selection import train_test_split    # Función utilizada para dividir un conjunto de datos en conjunto de entrenamiento y conjunto de prueba,
                                                        # lo cual será utilizado para ajustar el modelo a los datos y aprender lo patrones

from sklearn.preprocessing import StandardScaler    # Función utilizada para normalizar los datos, es decir, ajusta las características numéricas para
                                                    # que todas tengan una distribución estándar con media 0 y desviación estándar 1

                                                                            # `classification_report`da una visión general de cómo se está desempeñando el modelo
from sklearn.metrics import classification_report, roc_auc_score, f1_score  # `roc_auc_score` evalúa el área bajo la curva ROC (AUC-ROC), el rendimiento del modelo
                                                                            # `f1_score` es la media armónica entre la precisión y el recall


# Importar librería para el modelado de datos

                                    # `XGBClassifier` es una clase del paquete XGBoost que implementa un modelo de clasificación basado en el algoritmo `XGBoost`,
from xgboost import XGBClassifier   # el cual es un algoritmo muy poderoso y popular que combina varios árboles de decisión secuencialmente, donde cada nuevo árbol
                                    # corrige los errores del anterior, lo que le permite obtener un alto rendimiento en tareas de clasificación y regresión


# Importar librerías para el manejo de desbalance de datos

from imblearn.over_sampling import SMOTE    # `Synthetic Minority Over-sampling Technique` es una técnica de oversampling que se utiliza contra el desbalance en datasets
from collections import Counter     # `Counter` es una clase del módulo `collections`. Se utiliza para contar la cantidad de elementos en una secuencia o lista


# Importar librería para matriz de confusión

from sklearn import metrics     # Sirve para evaluar el rendimiento de modelos de machine learning. En este caso, para crear la matriz de confusión y su visualización.
```

### **7.2 Preprocesamiento De Datos**

```python
# Lista de columnas categóricas
columnas_categoricas = ['Trabajo', 'Estado Civil', 'Educacion', 'Incumplimiento', 'Vivienda', 'Prestamo', 'Contacto', 'Mes', 'Resultado Previo']

# `get_dummies()` es una función de pandas que se utiliza para convertir variables categóricas en una serie de columnas binarias (0 o 1)
# Este proceso se llama codificación one-hot o one-hot encoding
df_encoded = pd.get_dummies(data, columns=columnas_categoricas, drop_first=True)    # `drop_first=True` indica que se debe eliminar la primera
                                                                                # categoría de cada columna codificada para evitar la multicolinealidad
```

```python
# Eliminar la columna objetivo para crear un dataset con todas las características variables predictoras del modelo.
X = df_encoded.drop('Respuesta', axis=1)

# Convertir la columna objetivo a binaria ('yes' = 1, 'no' = 0) y crear un dataset únicamente con la variable objetivo, que es la etiqueta que se quiere predecir
y = df_encoded['Respuesta'].apply(lambda x: 1 if x == 'yes' else 0)
```

```python
# Normalizar variables numéricas

# Lista de columnas numéricas
columnas_numericas = ['Edad', 'Saldo', 'Duracion', 'Campaña', 'Dias Previos', 'Anterior', 'Dia']

# Crear una instancia de StandardScaler, al usarlo, las características se transforman para que todas tengan el mismo rango de importancia relativa
scaler = StandardScaler()

# Ajustar las columnas y transformarlas para que cada valor numérico se estandarice, de manera que todas tengan una media de 0 y desviación estándar de 1.
X[columnas_numericas] = scaler.fit_transform(X[columnas_numericas])
```

### **7.3 Entrenamiento y Prueba De Los Datos**

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
        # `test_size=0.3` indica que el 30% de los datos se utilizarán para el conjunto de prueba, y el 70% de los datos se utilizarán para el conjunto de entrenamiento
        # `random_state=42` controla la aleatoriedad del proceso de división, esto significa que cada vez que se ejecute código, la división será la misma
        # `stratify=y` asegura que la proporción 'y', (ejemplo, la cantidad de "yes" y "no") sea la misma tanto en el conjunto de entrenamiento como en el de prueba
```

### **7.4 Manejo Del Desbalance Por Medio De SMOTE**

```python
# Generar ejemplos sintéticos de la clase minoritaria en el conjunto de entrenamiento, equilibrando el conjunto de datos y mejorando la capacidad
# del modelo para aprender patrones de ambas clases. Nuevamente usando `random_state=42` para controlar la aletoriedad
smote = SMOTE(random_state=42)

# Aplicar SMOTE a los datos de entrenamiento, creando nuevos conjuntos de características y etiquetas equilibrados evitando el sesgo
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

# Contar y mostrar la distribución después de aplicar SMOTE, asegurando que ahora ambas clases están igualmente representadas en el conjunto de entrenamiento
print(f'Distribución de clases después de SMOTE: {Counter(y_train_balanced)}')
```

### **7.5 Entrenamiento Del Modelo XGBoost**

```python
# Crear una instancia del modelo XGBClassifier, que es parte de la biblioteca XGBoost
# Este modelo se utiliza para tareas de clasificación y es conocido por su eficacia y rendimiento, especialmente en datos desbalanceados
xgb_model = XGBClassifier(
    scale_pos_weight=len(y_train_balanced[y_train_balanced == 0]) / len(y_train_balanced[y_train_balanced == 1]),   # Ajusta la importancia de la minoría
    eval_metric='logloss',  # Es una métrica utilizada en problemas de clasificación, `logloss` toma en cuenta la confianza de las predicciones
    random_state=42     # Asegura que los resultados sean reproducibles
)

# `fit()` entrena el modelo xgb_model utilizando el conjunto de entrenamiento equilibrado (X_train_balanced y y_train_balanced) que se creó previamente
xgb_model.fit(X_train_balanced, y_train_balanced)
```

### **7.6 Evaluación Del Modelo**

```python
# Predicciones

# Esta función se utiliza para realizar predicciones en los datos de prueba (X_test)
y_pred = xgb_model.predict(X_test)

# Esta función devuelve las probabilidades predichas para cada clase
y_prob = xgb_model.predict_proba(X_test)[:, 1]


# Métricas

# Esta función calcula el AUC-ROC (Área Bajo la Curva de la Curva ROC). La curva ROC es una representación gráfica de la capacidad del modelo para clasificar
# correctamente las clases a diferentes umbrales de probabilidad. El AUC es una métrica que representa la capacidad del modelo para distinguir entre las dos clases
print(f'• AUC-ROC: {roc_auc_score(y_test, y_prob)}\n')

# Esta función calcula el F1-Score, que es una métrica que combina la precisión (precision) y la exhaustividad (recall) en un solo valor.
# Es útil cuando hay un desbalance entre las clases, ya que considera tanto los falsos positivos como los falsos negativos.
print(f'• F1-Score: {f1_score(y_test, y_pred)}\n')

# Esta función genera un informe que muestra las métricas de evaluación del modelo, incluyendo:
    # Precision: Proporción de verdaderos positivos sobre el total de positivos predichos.
    # Recall: Proporción de verdaderos positivos sobre el total de positivos reales.
    # F1-Score: Valor F1 para cada clase.
    # Support: Número de ocurrencias de cada clase en los datos de prueba
print(f'• Classification Report:\n\n{classification_report(y_test, y_pred)}')
```

### **7.7 Características Más Importantes**

```python
# Esta propiedad del modelo XGBClassifier devuelve la importancia de cada característica utilizada en el modelo
# La importancia se mide en función de cómo cada característica contribuye a las predicciones del modelo
xgb_feature_importance = pd.Series(xgb_model.feature_importances_, index=X.columns)

# Crear un gráfico de las 10 características más importantes en términos de su puntuación de importancia
xgb_feature_importance.nlargest(10).plot(kind='barh')
plt.title("Top 10 Características Más Importantes")
plt.show()
```

### **7.8 Matriz De Confusión**

```python
# Esta función calcula la matriz de confusión utilizando las etiquetas verdaderas "y_test" y las predicciones del modelo "y_pred"
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)

# Se utiliza para crear un objeto que facilita la visualización de la matriz de confusión
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=['False', 'True'])

# Crear y gráficar la visualización de la matriz de confusión
cm_display.plot()
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.show()
```
