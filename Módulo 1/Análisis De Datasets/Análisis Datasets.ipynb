# **ANÁLISIS DATASETS**

## **1. TIPOS DE DATOS**

Por medio de la función `select_dtypes()` es posible seleccionar los tipos de datos:


*   **Variables Cualitativas:**
  *   Cualitativas nominales: `include=['object']`.
  *   Cualitativas ordinales: `include=['category']`.
*   **Variables Cuantitativas:**
  *   Cuantitativas discretas: `include=['int']`.
  *   Cuantitativas continuas: `include=['float']`.

<br><br>
Luego de haber hecho lo anterior obtendremos un Dataframe según el tipo de dato especificado, por lo que por medio de `.columns` obtendremos las columnas con los tipos de datos y gracias a `to_list()` los datos estarán ordenados en una lista.

**DATASET 1**

```python
nominales = dataset_1.select_dtypes(include=['object']).columns.to_list()
discretas = dataset_1.select_dtypes(include=['int']).columns.to_list()
continuas = dataset_1.select_dtypes(include=['float']).columns.to_list()

print(f'• Variables cualitativas nominales: {nominales[0]}.')
print(f'• Variables cuantitativas discretas: {discretas[0]} y {discretas[1]}.')
print(f'• Variables cuantitativas continuas: {continuas[0]} y {continuas[1]}.')
```

**DATASET 2**

```python
discretas = dataset_2.select_dtypes(include=['int']).columns.to_list()
continuas = dataset_2.select_dtypes(include=['float']).columns.to_list()

print(f'• Variables cuantitativas discretas: {discretas[0]} y {discretas[1]}.')
print(f'• Variables cuantitativas continuas: {continuas[0]} y {continuas[1]}.')
```

**DATASET 3**

```python
nominales = dataset_3.select_dtypes(include=['object']).columns.to_list()
discretas = dataset_3.select_dtypes(include=['int']).columns.to_list()
continuas = dataset_3.select_dtypes(include=['float']).columns.to_list()

print(f'• Variables cualitativas nominales: {nominales[0]}.')
print(f'• Variables cuantitativas discretas: {discretas[0]}.')
print(f'• Variables cuantitativas continuas: {continuas[0]}, {continuas[1]}, {continuas[2]} e {continuas[3]}.')
```

## **2. CANTIDAD DE DATOS**

Al utilizar `shape` obtendremos una tupla con las $filas*columnas$ del dataframe y por medio de la función `count()` obtendremos un _Series_ con la cantidad de datos por columnas, por lo que al combinarlo con la función `sum()` obtendremos la totalidad de datos sumando cada dato por columna.

**DATASET 1**

```python
print(f'El dataset tiene {dataset_1.shape[0]} filas y {dataset_1.shape[1]} columnas. Con un total de {dataset_1.count().sum()} datos.')
```

**DATASET 2**

```python
print(f'El dataset tiene {dataset_2.shape[0]} filas y {dataset_2.shape[1]} columnas. Con un total de {dataset_2.count().sum()} datos.')
```

**DATASET 3**

```python
print(f'El dataset tiene {dataset_3.shape[0]} filas y {dataset_3.shape[1]} columnas. Con un total de {dataset_3.count().sum()} datos.')
```

## **3. PATRONES**

Por medio de la función `histplot()` es posible visualizar la distribución de los datos con respecto a la columna del Dataframe, de esta forma es posible encontrar patrones en los datos.

**DATASET 1**

```python
columnas = dataset_1.columns[1:]

fig, axes = plt.subplots(nrows=1, ncols=len(columnas), figsize=(15, 4))

for i, columna in enumerate(columnas):
    sns.histplot(dataset_1[columna], kde=True, ax=axes[i])
    axes[i].set_title(columna)

plt.tight_layout()
plt.show()
```

**DATASET 2**

```python
columnas = dataset_2.columns[1:]

fig, axes = plt.subplots(nrows=1, ncols=len(columnas), figsize=(15, 4))

for i, columna in enumerate(columnas):
    sns.histplot(dataset_2[columna], kde=True, ax=axes[i])
    axes[i].set_title(columna)

plt.tight_layout()
plt.show()
```

**DATASET 3**

```python
columnas = dataset_3.columns[1:]

fig, axes = plt.subplots(nrows=1, ncols=len(columnas), figsize=(15, 4))

for i, columna in enumerate(columnas):
    sns.histplot(dataset_3[columna], kde=True, ax=axes[i])
    axes[i].set_title(columna)

plt.tight_layout()
plt.show()
```

## **4. DATOS ATÍPICOS**

Por medio de la función `boxplot()` es posible encontrar _outliers_, es decir, datos atípicos, viendo cómo están distribuidos los datos y viendo cómo algunos de estos están por fuera de los cuartiles.

**DATASET 1**

```python
columnas = dataset_1.select_dtypes(include=['int', 'float']).columns[1:]

fig, axes = plt.subplots(nrows=1, ncols=len(columnas), figsize=(15, 4))

for i, columna in enumerate(columnas):
    dataset_1.boxplot(columna, ax=axes[i])
    axes[i].set_title(columna)

plt.tight_layout()
plt.show()
```

**DATASET 2**

```python
columnas = dataset_2.select_dtypes(include=['int', 'float']).columns[1:]

fig, axes = plt.subplots(nrows=1, ncols=len(columnas), figsize=(15, 4))

for i, columna in enumerate(columnas):
    dataset_2.boxplot(columna, ax=axes[i])
    axes[i].set_title(columna)

plt.tight_layout()
plt.show()
```

**DATASET 3**

```python
columnas = dataset_3.select_dtypes(include=['int', 'float']).columns[1:]

fig, axes = plt.subplots(nrows=1, ncols=len(columnas), figsize=(15, 4))

for i, columna in enumerate(columnas):
    dataset_3.boxplot(columna, ax=axes[i])
    axes[i].set_title(columna)

plt.tight_layout()
plt.show()
```

## **5. DATOS FALTANTES**

Por medio de la función `isna()` obtenemos el Dataframe pero con booleanos _True_ y _False_ si la celda está vacía o no correspodientemente, y así como _True_ también equivale a 1 y _False_ a 0, combinando esa función con `sum()` obtendremos un _Series_ de una suma de datos <font color='#5292c7'>**NaN**</font> o <font color='#5292c7'>**None**</font> por columna, finalmente si volvemos a usar la función `sum()` a ese resultado obtendremos la suma de los datos faltantes en todo el Dataframe, si por el contrario al final usamos `any()` obtendremos un booleano en el que retornará _True_ si encuentra por lo menos un solo dato <font color='#5292c7'>**NaN**</font> o <font color='#5292c7'>**None**</font> y _False_ si el Dataframe está completo.

**DATASET 1**

```python
print(f'Hay {dataset_1.isna().sum().sum()} datos faltantes.' if dataset_1.isna().sum().any() else 'No hay datos faltantes.')
```

**DATASET 2**

```python
print(f'Hay {dataset_2.isna().sum().sum()} datos faltantes.' if dataset_2.isna().sum().any() else 'No hay datos faltantes.')
```

**DATASET 3**

```python
print(f'Hay {dataset_3.isna().sum().sum()} datos faltantes.' if dataset_3.isna().sum().any() else 'No hay datos faltantes.')
```
